```python
og = plt.imread("occlusion_study_results_all_3/occlusion_nlx_gpt_results/ILSVRC2012_val_00001350/ILSVRC2012_val_00001350.JPEG")
blocked = plt.imread("occlusion_study_results_all_3/occlusion_nlx_gpt_results/ILSVRC2012_val_00001350/ILSVRC2012_val_00001350_x.jpg")
blocked = cv2.cvtColor(blocked, cv2.COLOR_BGR2GRAY)
```

```python
or_image = np.where(blocked == 255, blocked, 0)
kernel = np.ones((3, 3), dtype=np.uint8)
opened_image = cv2.morphologyEx(or_image, cv2.MORPH_OPEN, kernel)
kernel = np.ones((13, 13), dtype=np.uint8)
dilated_image = cv2.morphologyEx(opened_image, cv2.MORPH_DILATE, kernel)
```

```python
plt.imshow(np.where(np.dstack([dilated_image] * 3) == 255, og, 0))
```
